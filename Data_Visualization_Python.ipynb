# Line Charts and Time Series:

```python
#Compare GDP per Capita in USA and China
usa_gdp = data[data.country=='United States']
```

```python
usa_gdp.head()
```

```python
china_gdp = data[data.country=='China']
china_gdp.head()
```

```python
plt.plot(usa_gdp.year,usa_gdp.gdpPerCapita)
plt.xlabel('Years')
plt.ylabel('GDP per Capita')
plt.plot(china_gdp.year,china_gdp.gdpPerCapita)
plt.legend(['USA','China'])
plt.show()
```

This is a little difficult to compare since the absolute values are pretty different.

```python
usa_growth = usa_gdp.gdpPerCapita/usa_gdp.gdpPerCapita.iloc[0]*100
china_growth = china_gdp.gdpPerCapita/china_gdp.gdpPerCapita.iloc[0]*100
```

```python
plt.plot(usa_gdp.year,usa_growth)
plt.xlabel('Years')
plt.ylabel('GDP per Capita')
plt.plot(china_gdp.year,china_growth)
plt.legend(['USA','China'])
plt.show()
```

This shows that GDP of China has grown relatively faster.

```python
#Comparing population growth in USA and China
plt.plot(usa_gdp.year,usa_gdp.population)
plt.xlabel('Years')
plt.ylabel('Population')
plt.plot(china_gdp.year,china_gdp.population)
plt.legend(['USA','China'])
plt.show()
```

```python
plt.plot(usa_gdp.year,usa_gdp.population/10**6)
plt.xlabel('Years')
plt.ylabel('Population')
plt.plot(china_gdp.year,china_gdp.population/10**6)
plt.legend(['USA','China'])
plt.show()
```

```python
usa_growth = usa_gdp.population/usa_gdp.population.iloc[0]*100
china_growth = china_gdp.population/china_gdp.population.iloc[0]*100
```

```python
plt.plot(usa_gdp.year,usa_growth)
plt.xlabel('Years')
plt.ylabel('Population')
plt.plot(china_gdp.year,china_growth)
plt.legend(['USA','China'])
plt.show()
```

We can conclude, that the population growth rate was similar till 1960, after which China's population grew a lot more than USA.

# Scatter Plots:


```python
#Compare life expectancy and GDP
import numpy as np
```

```python
data_2007=data[data.year==2007]
```

```python
plt.scatter(data_2007.gdpPerCapita,data_2007.lifeExpectancy,5) #5 is the size
plt.title('2007 Data')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```

Higher the GDP, higher is the life expectancy.
Also, initially as GDP increases, life expectancy increases exponentially.
However after GDP has reached a certain value, an increase does not have much effect on life expectancy.

```python
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy)
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita),data_2007.lifeExpectancy,10) #10 is the size
plt.title('2007 Data')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
years_sorted = sorted(set(data.year))
years_sorted
```

```python
for given_year in years_sorted:
    data_year = data[data.year==given_year]
    plt.scatter(data_year.gdpPerCapita,data_year.lifeExpectancy,10) #10 is the size
    plt.xlim(0,60000)
    plt.ylim(25,85)
    plt.title(given_year)
    plt.xlabel('GDP per Capita')
    plt.ylabel('Life Expectancy')
    #plt.show()
    plt.savefig(str(given_year),dpi=200)
    plt.clf()
```

# Bar graphs:

```python
#Compare populations of 10 most populous countries in 2007
data_2007 = data[data.year==2007]
data_2007.head()

```

```python
data_2007.sort_values('population', ascending = False).head(10)
```

```python
data_2007_top_ten = data_2007.sort_values('population', ascending = False).head(10)
```

```python
x = range(10)
plt.bar(x,data_2007_top_ten.population/10**6)
plt.title('Population')
plt.xticks(x,data_2007_top_ten.country,rotation='vertical')
plt.ylabel('Population in millions')
plt.show()
```

```python
#Compare GDP of 10 most populous countries
gdp= data_2007_top_ten.population * data_2007_top_ten.gdpPerCapita
x=range(10)
plt.bar(x,gdp/10**9)
plt.title('GDP')
plt.xticks(x,data_2007_top_ten.country,rotation='vertical')
plt.ylabel('GDP in Billions')
plt.show()
```

```python
gdp= data_2007_top_ten.population * data_2007_top_ten.gdpPerCapita
x=range(10)
plt.subplot(211)
plt.bar(x,data_2007_top_ten.population/10**6)

plt.xticks([],[])
plt.ylabel('Population in millions')
plt.subplot(212)
plt.bar(x,gdp/10**9)

plt.xticks(x,data_2007_top_ten.country,rotation='vertical')
plt.ylabel('GDP in Billions')
plt.show()
```

```python
data_2007 = data[data.year == 2007]
top10 = data_2007.sort_values('population', ascending=False).head(10)
```

```python
x = np.arange(10)
fig, ax1 = plt.subplots()
width = 0.3 # This is the width of each bar in the bar plot.
plt.xticks(x, top10.country, rotation='vertical')
population = ax1.bar(x, top10.population / 10**6, width)
plt.ylabel('Population')
# ax1.twinx() gives us the same x-axis with the y-axis on the right.
ax2 = ax1.twinx()
gdp = ax2.bar(x + width, top10.gdpPerCapita * top10.population / 10**9,
              width, color='orange')
plt.ylabel('GDP')
plt.legend([population, gdp],
           ['Population in Millions', 'GDP in Billions'])
figure = plt.gcf() # get current figure
plt.show()
```

# Data Aggregation:


```python
data = pd.read_csv('obama.csv',parse_dates=['year_month'])
#parse_dates tells pandas to interpret year_month as time stamp
```

```python
data
```

```python
plt.plot(data.year_month,data.approve_percent,'o',markersize=2,alpha=0.3)
plt.show()
```

```python
data_mean=data.groupby('year_month').mean()
```

```python
data_mean
```

```python
plt.plot(data_mean.index,data_mean.approve_percent,'red')
plt.plot(data.year_month,data.approve_percent,'o',markersize=2,alpha=0.3)
plt.xlabel('Years')
plt.ylabel('Approval Percentage')
plt.show()
```
