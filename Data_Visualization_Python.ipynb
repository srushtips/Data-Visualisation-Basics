```python
countries = pd.read_csv('countries.csv')
```

```python
countries
```

```python
china = countries[countries.country =='China']
```

```python
china
```

```python
usa = countries[countries.country=='United States']
```

```python
usa
```

```python
plt.plot(usa.year,usa.population)
plt.title('Population vs Year')
plt.xlabel('Years')
plt.ylabel('Population')
plt.show()
```

However, Y axis has exponential data. Hence, converting it to a smaller value.

```python
plt.plot(usa.year,usa.population/10000)
plt.title('Population vs Year')
plt.xlabel('Years')
plt.ylabel('Population/10000')
plt.show()
```

Now, let's compare the growth of population in China and USA.

```python
plt.plot(usa.year,usa.population/10000)
plt.plot(usa.year,china.population/10000)
plt.title('Population vs Year')
plt.xlabel('Years')
plt.ylabel('Population/10000')
plt.legend(['USA','China'])
plt.show()
```

We can conclude that China began with a higher population and also has had a higher growth in population.
